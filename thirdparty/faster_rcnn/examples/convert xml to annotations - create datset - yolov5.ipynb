```python
from xml.dom import minidom
import os


### Edit the following line for input
f= open("annotations.txt","w+")

img_list_test = list()
ano = './Annotations/'
img = './Images/'
dfiles = os.listdir(ano)
j=0
for files in dfiles:
    #print(os.path.join(ano,files))
    x=[0,0,0,0]
    y=[0,0,0,0]

    ##ADDED BY ESHAAN BECAUSE THIS FILE WAS CAUSING TROUBLE 
    if (files == ".DS_Store"):
        continue;
    
    mydoc = minidom.parse(os.path.join(ano,files))
    clas=mydoc.getElementsByTagName('name')
    i=0
    for cl in clas:
        c=cl.firstChild.data
        polygon = mydoc.getElementsByTagName('polygon')

        x[0]=int(polygon[i].childNodes[1].childNodes[0].firstChild.data)
        y[0]=int(polygon[i].childNodes[1].childNodes[1].firstChild.data)

        x[1]=int(polygon[i].childNodes[2].childNodes[0].firstChild.data)
        y[1]=int(polygon[i].childNodes[2].childNodes[1].firstChild.data)

        x[2]=int(polygon[i].childNodes[3].childNodes[0].firstChild.data)
        y[2]=int(polygon[i].childNodes[3].childNodes[1].firstChild.data)

        x[3]=int(polygon[i].childNodes[4].childNodes[0].firstChild.data)
        y[3]=int(polygon[i].childNodes[4].childNodes[1].firstChild.data)
        i+=1
        
        xmax=max(x)
        ymax=max(y)
        xmin=min(x)
        ymin=min(y)

        path_var = os.path.join("Images",files)
        path_var = path_var[:-3] + "jpg"
        
        print_var = path_var +","+str(xmin)+","+str(ymin)+","+str(xmax)+","+str(ymax)+","+c

        print(print_var)
        f.write(print_var+"\n")
#         img_list_test.append(path_var +","+str(xmin)+","+str(ymin)+","+str(xmax)+","+str(ymax)+","+c)
        j+=1
f.close()
print(j)
```

```python
### Generate the Train, Test, and Validation dataset files
```

```python
import csv
import cv2
import random

file_list = list()

with open('annotations.txt', 'r') as csvfile:
    sample = csv.reader(csvfile, delimiter=',')
    all_images = list()
    i = 0
    for row in sample:
        print(i, row)
        i+=1
        str1 = row[0] + "," + row[1] + "," + row[2] + "," + row[3] + "," + row[4] + "," + row[5]
        all_images.append(str1)
        file_list.append(row[0])

file_list = list(set(file_list))

print(len(all_images))
print(len(file_list))

random.shuffle(file_list)

train = list()
test = list()
val = list()

# random.shuffle(all_images)

train_len = int(0.8 * len(file_list))
test_len = int(0.9 * len(file_list))
train_file = file_list[:train_len]
test_file = file_list[train_len:test_len]
val_file = file_list[test_len:]

f1=open("train_labels.txt", "a+")
f2=open("test_labels.txt", "a+")
f3=open("val_labels.txt", "a+")

for i in range(len(all_images)):
    print ('r: ', r)
    r = all_images[i]
    fl = r.split(',')[0]
    print ('fl: ',fl, 'r[0]:',r[0])
    if fl in train_file:
        f1.write(str(r)+'\n')
    elif fl in test_file:
        f2.write(str(r)+'\n')
    else:
        f3.write(str(r)+'\n')
    
f1.close()

f2.close()

f3.close()

print("Done")
```

```python
import csv
import cv2
with open('annotations_6_17_21.txt', newline='') as csvfile:
    sample = csv.reader(csvfile, delimiter=',')
#     print(type(sample))
#     counter = 0
    for row in sample:
#         counter += 1
#         if counter>5:
#             break
        img = row[0]
        im = cv2.imread(str(img))
        height = im.shape[0]
        print(type(height))
        width = im.shape[1]
        bounding_box = row[1:-1]
        new_b1 = (int(bounding_box[0])+int(bounding_box[2]))/(2*width)
        new_b2 = (int(bounding_box[1])+int(bounding_box[3]))/(2*height)
        new_b3 = (int(bounding_box[2])-int(bounding_box[0]))/(width)
        new_b4 = (int(bounding_box[3])-int(bounding_box[1]))/(height)
        label = 0 # default : blemished
        if row[-1] == 'b':
            label = 0
        elif row[-1] == 'u':
            label = 1
        file_name = img.split('.jpg')[0] + '.txt'
        fl = open(file_name,"a+")
        out_now = str(label)+' '+str(new_b1)+' '+str(new_b2)+' '+str(new_b3)+' '+str(new_b4)+'\n'
        fl.write(out_now)
        fl.close()
        print(out_now)
        print(file_name)
print("Done")
```
